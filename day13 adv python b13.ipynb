```python
import numpy as np

import pandas as pd

from matplotlib import pyplot as plt
```

```python
df =pd.read_csv('countries (2).csv')
df
```

```python
us = df[df.country =='United States']

us
```

```python
ch = df[df.country =='China']

ch
```

```python
#COMPARISION OF CHINA & US
```

```python
plt.plot(us.year,us.population)

plt.show()
```

```python
plt.plot(us.year,us.population/10**6)

plt.show()
```

```python
plt.plot(us.year,us.population/10**6)

plt.plot(ch.year,ch.population/10**6)

plt.title('comparision of us and china population')

plt.xlabel('year')

plt.ylabel('population')

plt.legend(['this is us data','this is china data'])

plt.show()
```
